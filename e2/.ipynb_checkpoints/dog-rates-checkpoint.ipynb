```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('./dog_rates_tweets.csv', parse_dates=['created_at'])
```

```python
# df['created_at'] = pd.to_datetime(df['created_at'])
type(df['created_at'][0])
```

```python
 df['rating'] = df['text'].str.extract(r'(\d+(\.\d+)?)/10')[0].astype('float')
```

```python
df[df.rating.notnull()]
```

```python
df_sorted = df[df['rating'] <= 25]
```

```python
df_sorted['timestamp'] = df['created_at'].apply(lambda x: x.timestamp())
```

```python
fit = stats.linregress(df_sorted['timestamp'].values, df_sorted['rating'].values)
```

```python
fit.slope, fit.intercept
```

```python
y = fit.slope*df_sorted['created_at']+fit.intercept 
```

```python
plt.xticks(rotation=25)
plt.plot(df_sorted['created_at'].values, df_sorted['rating'].values, 'b.', alpha=0.5)
plt.plot(df_sorted['created_at'].values, y, 'r-', linewidth=3)
```
